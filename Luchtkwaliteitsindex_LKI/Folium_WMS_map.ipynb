```python
import folium
from branca.element import Element

# Create the map
m = folium.Map(location=[52.2, 4.2], zoom_start=7.5, tiles="cartodb positron")

# Add WMS tile layer
folium.WmsTileLayer(
    url="https://data.rivm.nl/geo/lucht/wms",
    layers="actueel_lki",
    name="PM10",
    fmt="image/png",
    overlay=True,
    transparent=True,
    control=True,
    show=True
).add_to(m)

# Legend URL
legend_url = "https://data.rivm.nl/geo/lucht/ows?service=WMS&version=1.3.0&request=GetLegendGraphic&format=image%2Fpng&width=20&height=20&layer=actueel_lki"

# Legend HTML with refined box-shadow
legend_html = f"""
<div style="
    position: fixed;
    bottom: 50px;
    left: 50px;
    width: 200px;  /* Adjusted width */
    height: auto;
    background-color: white;  /* Add background for contrast */
    padding: 5px;  /* Add padding for better spacing */
    border: 1px solid #ccc;  /* Add a light border */
    box-shadow: 3px 3px 8px rgba(0, 0, 0, 0.3);  /* Subtle shadow */
    z-index:9999;
">
    <img src="{legend_url}" style="width: 100%; height: auto;">
</div>
"""

# Add legend to the map
legend = Element(legend_html)
m.get_root().html.add_child(legend)

# Add layer control
folium.LayerControl().add_to(m)

# Display the map
m
```
