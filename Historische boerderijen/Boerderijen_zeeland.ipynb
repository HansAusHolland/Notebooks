```python
import geopandas as gpd
import matplotlib.pyplot as plt
import folium

map_pad = "https://opengeodata.zeeland.nl/geoserver/ovz/wfs?request=GetFeature&service=WFS&version=1.1.0&outputFormat=application%2Fjson&typeName=ovz:ext_ovz_xyz_geocmd_hisboepnt"

gdf = gpd.read_file(map_pad)
gdf = gdf.to_crs(epsg = 4326)
m = folium.Map(location=[51.5, 3.9],tiles="OpenStreetMap", zoom_start=10)

for _, row in gdf.iterrows():
    coords = [row.geometry.y, row.geometry.x]

    html = f"""
    <div>
        <img src="{row["ol_foto1"]}" alt="Foto" width="300" height="200">
    </div>
    """
    iframe = folium.IFrame(html, width=320, height=240)
    popup = folium.Popup(iframe, max_width=320)

    folium.CircleMarker(
        radius = 5,
        color="green",
        fill=True,
        location=coords,
        popup=popup,
    ).add_to(m)
#m.save("map_with_photos.html")
m
```

```python

import geopandas as gpd
import matplotlib.pyplot as plt
import folium

map_pad = "/Users/hbraakmann/Desktop/Geopandas/Boerderijen_zeeland/ext_ovz_xyz_geocmd_hisboepnt/ext_ovz_xyz_geocmd_hisboepnt.shp"

gdf = gpd.read_file(map_pad)
gdf = gdf.to_crs(epsg = 4326)
m = folium.Map(location=[51.5, 3.9],tiles="OpenStreetMap", zoom_start=10)

icon_image = "https://cdn-icons-png.flaticon.com/256/4661/4661507.png"
    
for _, row in gdf.iterrows():
    coords = [row.geometry.y, row.geometry.x]
    
    html = f"""
    <div>
        <p>{row["straat"]} , {row["plaats"]}</p>
        <img src="{row["ol_foto1"]}" alt={row["straat"]} width="300" height="200">
    </div>
    """
    
    iframe = folium.IFrame(html, width=320, height=240)
    popup = folium.Popup(iframe, max_width=320)
    folium.Marker(
        location = coords,
        icon=folium.CustomIcon(icon_image, icon_size=(30,30)),
        popup=popup,
    ).add_to(m)
#m.save("boerderijen_zeelend.html")
m
```
