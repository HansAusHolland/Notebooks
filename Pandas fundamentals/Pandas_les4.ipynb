```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
snoep_shop= {
    "Snoep":["drop", "lollies", "appel", "spekkies", "zuurtjes", "bonbons"],
    "Kilo's verkocht":[3, 4.5, 12, 7, 1.5, 6],
    "Prijs per kg":[1.99, 2.00, 0.50, 2.99, 1.65, 5.00],
}
```

```python
df = pd.DataFrame(snoep_shop).round(2)
```

```python
df["Prijs per kg"]>3
```

```python
prijs_sel = df.loc[df["Prijs per kg"]>3]
prijs_sel
```

```python
df["Totale opbrengst"] = round(df["Kilo's verkocht"] * df["Prijs per kg"],2)
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df["Snoep"], df["Totale opbrengst"], color="skyblue", linewidth=1, edgecolor="black")

#Titel en labels voor x-as en y-as
plt.title("Snoepwinkel Opbrengst in €")
plt.xlabel("Snoep", fontsize=14)
plt.ylabel("Opbrengst", fontsize=14)

# Formateren van de tekst op de x-as
plt.xticks(rotation=45, ha="right")

# Grid lijnen voor de grafiek
plt.grid(axis="y", linestyle="--")

#Legenda
plt.legend(["Totale opbrengst"])

#Opmaak
plt.tight_layout()
plt.show()
```
