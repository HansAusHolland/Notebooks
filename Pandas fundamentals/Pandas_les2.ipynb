# Les 2: Geavanceerde DataFrame Manipulaties met Pandas

Welkom terug! Vandaag gaan we dieper ingaan op enkele geavanceerde technieken voor het manipuleren van DataFrames in Pandas. We zullen de technieken `groupby`, pivot-tabellen en het samenvoegen van DataFrames verkennen, met behulp van een dataset over Spotify-artiesten.

## Stap 1: Importeren van Pandas

Eerst moeten we de Pandas-bibliotheek importeren. Pandas helpt ons om gemakkelijk met gegevens te werken.


```python
import pandas as pd
```

## Stap 2: Onze gegevens

Laten we een dataset creëren over verschillende artiesten en hun nummers op Spotify. Deze dataset bevat informatie over de naam van de artiest, de titel van het nummer, het aantal streams en de prijs per stream.

```python
# Onze gegevens
spotify_artiesten = {
    "Artiest": ["Drake", "Taylor Swift", "Ed Sheeran", "Billie Eilish", "BTS", "Ariana Grande"],
    "Titel": ["God's Plan", "Love Story", "Shape of You", "Bad Guy", "Dynamite", "7 rings"],
    "Streams": [1500000000, 300000000, 2600000000, 1900000000, 1700000000, 2100000000],
    "Prijs per stream": [0.004, 0.004, 0.004, 0.004, 0.004, 0.004],
}

# Omzetten van de dictionary naar een Pandas DataFrame
spotify_df = pd.DataFrame(spotify_artiesten)

# Bekijken van de eerste paar rijen van de DataFrame
spotify_df.head()
```

## Stap 3: Verkennen van de DataFrame

Laten we eens kijken naar de structuur van onze DataFrame en de eerste paar rijen bekijken.

```python
# Bekijk de structuur van de DataFrame
spotify_df.info()
```

    </ul>
  </div>
</details>

## Stap 4: Eenvoudige bewerkingen

Laten we nu enkele eenvoudige bewerkingen uitvoeren op onze DataFrame. We beginnen met het berekenen van het totale aantal streams.

```python
# Totale hoeveelheid streams
totaal_streams = spotify_df["Streams"].sum()
print(f"Totaal aantal streams: {totaal_streams}")
```

<details>
  <summary style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px; cursor: pointer;"><strong>Tip!</strong></summary>
  <div style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px;">
    <ul>
      <li><code>spotify_df["Streams"]</code> selecteert de kolom met het aantal streams.</li>
      <li><code>.sum()</code> berekent de som van alle waarden in deze kolom.</li>
    </ul>
  </div>
</details>

## Stap 5: Nieuwe kolom toevoegen

We gaan een nieuwe kolom toevoegen die de totale inkomsten voor elk nummer berekent (streams * prijs per stream).

```python
# Nieuwe kolom 'Totale inkomsten' toevoegen
spotify_df['Totale inkomsten'] = spotify_df["Streams"] * spotify_df['Prijs per stream']

# Bekijken van de bijgewerkte DataFrame
spotify_df.head()
```

<details>
  <summary style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px; cursor: pointer;"><strong>Tip!</strong></summary>
  <div style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px;">
    <ul>
      <li>We vermenigvuldigen de waarden in de kolom "Streams" met de waarden in de kolom "Prijs per stream".</li>
      <li>De resultaten worden opgeslagen in een nieuwe kolom genaamd 'Totale inkomsten'.</li>
    </ul>
  </div>
</details>

## Stap 6: Groeperen en aggregeren

Laten we de gegevens groeperen op basis van de artiest en de totale streams per artiest berekenen.


```python
# Groeperen op 'Artiest' en de som van de streams berekenen
groepering = spotify_df.groupby('Artiest')["Streams"].sum()
groepering
```

<details>
  <summary style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px; cursor: pointer;"><strong>Tip!</strong></summary>
  <div style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px;">
    <ul>
      <li><code>groupby('Artiest')</code> groepeert de rijen van de DataFrame op basis van de waarden in de kolom 'Artiest'.</li>
      <li><code>sum()</code> berekent de som van de "Streams" binnen elke groep artiesten.</li>
    </ul>
  </div>
</details>

## Stap 7: Pivot-tabellen

We maken een pivot-tabel om de totale inkomsten voor elke artiest te bekijken.

```python
# Maak een pivot-tabel
pivot_tabel = spotify_df.pivot_table(values='Totale inkomsten', index='Artiest', aggfunc='sum')
pivot_tabel
```

<details>
  <summary style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px; cursor: pointer;"><strong>Tip!</strong></summary>
  <div style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px;">
    <ul>
      <li><code>pivot_table</code> maakt een nieuwe tabel (pivot-tabel) waarbij de rijen worden gegroepeerd op basis van de waarde in de kolom 'Artiest'.</li>
      <li><code>values='Totale inkomsten'</code> specificeert de kolom waarvan we de waarden willen gebruiken.</li>
      <li><code>aggfunc='sum'</code> specificeert dat we de waarden willen sommeren.</li>
    </ul>
  </div>
</details>

## Stap 8: DataFrames samenvoegen

Stel dat we een tweede dataset hebben met informatie over de muzieklabels van de artiesten. We gaan deze datasets samenvoegen.

```python
# Tweede dataset met informatie over muzieklabels
muzieklabels = {
    "Artiest": ["Drake", "Taylor Swift", "Ed Sheeran", "Billie Eilish", "BTS", "Ariana Grande"],
    "Label": ["Universal", "Republic", "Atlantic", "Interscope", "Big Hit", "Republic"]
}

labels_df = pd.DataFrame(muzieklabels)

# Samenvoegen van de twee DataFrames
samengevoegd_df = pd.merge(spotify_df, labels_df, on="Artiest")
samengevoegd_df
```

<details>
  <summary style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px; cursor: pointer;"><strong>Tip!</strong></summary>
  <div style="background-color: #f0f8ff; border-left: 6px solid #1e90ff; padding: 10px;">
    <ul>
      <li>We creëren een nieuwe DataFrame <code>labels_df</code> met informatie over de muzieklabels van elk type artiest.</li>
      <li><code>pd.merge(spotify_df, labels_df, on="Artiest")</code> voegt de twee DataFrames samen op basis van de gemeenschappelijke kolom 'Artiest'.</li>
    </ul>
  </div>
</details>
